```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d

# control step
TIME_STEP = 64

# number of distance sensors
N = 8

# motor speed unit
SPEED_UNIT = 0.00628

# for collision avoidance
braitenberg_matrix = np.array([
    [150, -35],
    [100, -15],
    [80, -10],
    [-10, -10],
    [-10, -10],
    [-10, 80],
    [-30, 100],
    [-20, 150]])

# edge detection matrix
conv_matrix = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]])

# create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# avoid dummy values on first device query
step = 0
samples = 0

while True:
    step += 1

    # read all distance sensors
    sensor_values = np.random.rand(N) * 2000  # replace with actual sensor values

    # get camera RGB image (replace with actual image)
    rgb = np.random.randint(0, 255, size=(64, 64, 3), dtype=np.uint8)

    # display inversed rgb image
    inverse = 255 - rgb
    axs[0, 0].imshow(inverse)
    axs[0, 0].set_title('RGB Camera')

    # create intensity image from RGB image
    intens = np.mean(rgb, axis=2)

    # edge detection
    edges = convolve2d(intens, conv_matrix, mode='valid')

    # display 'canny' image
    axs[0, 1].imshow(edges, cmap='gray')
    axs[0, 1].set_title('Edge detection')

    # plot distance sensor values
    axs[1, 1].bar(np.arange(1, N + 1), sensor_values)
    axs[1, 1].set_title('Distance sensors')
    axs[1, 1].set_ylim(0, 10)

    # braitenberg collision avoidance
    speed = (1 - (sensor_values / 500)) @ braitenberg_matrix * 5

    # avoid Webots warning
    speed = np.minimum(speed, 1000)

    # plot trajectory (replace with actual trajectory)
    axs[1, 0].plot(np.random.rand(100), np.random.rand(100), 'b-')
    axs[1, 0].plot(np.random.rand(1), np.random.rand(1), 'ro')
    axs[1, 0].set_title('Trajectory (Supervisor)')

    # adjust layout and show plot
    plt.tight_layout()
    plt.draw()
    plt.pause(0.01)

```

2. **Edge detection**: Gambar ini menampilkan hasil deteksi tepi pada citra intensitas (rata-rata dari saluran RGB). Deteksi tepi dilakukan dengan konvolusi citra intensitas dengan matriks deteksi tepi.
