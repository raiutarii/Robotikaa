```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat data waktu dan sudut
time = np.linspace(0, 10, 500)
angle = np.sin(time)

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(time, angle)
plt.title('Plot Sudut vs Waktu')
plt.xlabel('Waktu (s)')
plt.ylabel('Sudut (rad)')

# Menyimpan plot ke file
plt.savefig('plot.png')

# Menyimpan data ke file teks
np.savetxt('plot.txt', np.column_stack([time, angle]))

# Menampilkan plot
plt.show()
```
